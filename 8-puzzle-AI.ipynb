```python
import copy
import math
import heapq
import time
import numpy as np

k=9 #3*3
kk=3 #3*3
goal = "012345678"

def swap(state, x,y):
    returnedState=''
    for i in range(0,k):
      ch = state[i]
      if ch == state[x]:
        ch = state[y]
      elif ch == state[y]:
        ch = state[x]
      returnedState = returnedState + ch
    return returnedState

def create_child(state,explored):
  zeroIndex = state.find('0')
  child = []
  if zeroIndex / kk >= 1:       #UP
    child.append(swap(state,zeroIndex,zeroIndex-kk))
  if zeroIndex % kk != 0:  #left
    child.append(swap(state,zeroIndex,zeroIndex-1))
  if ((k-1)-zeroIndex >= kk ):   #down
    child.append(swap(state,zeroIndex,zeroIndex+kk))
  if (zeroIndex-(kk-1))%kk != 0:   #Right
    child.append(swap(state,zeroIndex,zeroIndex+1))
  explored.add(state)                    
  return child

######################################## 
def changer(s):
  m=np.zeros((3,3))
  ind=0
  for i in range(0,3):
    if(ind>8): break
    for j in range(0,3):
      if(ind>8): break
      m[i][j]=int(s[ind])
      ind+=1
  return m

def find_index(matrix,var):
    for row in range(0,3):
      for column in range(0,3):
            if matrix[row][column] == var:
                return (row,column)

def manhattan(puzzle,goal):
  cost =0
  for i in range(0,3):
    for j in range(0,3):
      if (puzzle[i][j]!=0):
        x,y=find_index(goal,puzzle[i][j])
        cost += abs(x-i)+abs(y-j)
  return cost

def ED(puzzle,goal):
  cost =0
  for i in range(0,3):
    for j in range(0,3):
      if (puzzle[i][j]!=0):
        x,y=find_index(goal,puzzle[i][j])
        cost += math.sqrt((x-i)**2+(y-j)**2)
  return cost

def inversions(s):
        count = 0
        for i in range(len(s)-1):
            for j in range(i+1,len(s)):
                if s[i] > s[j]:
                    count +=  1
                    break
        return (count%2==0)

def getMaxDepth(parent):
  costmax=0
  for i in range(0,len(parent)):
       p,cost=list(parent.values())[i]
       if(costmax<cost): costmax = cost
  return costmax 
#################################################################
def aStarPath2Goal(initialState,parent,endPoint):
  path = [goal]
  while path[-1] != initialState:
    path.append(parent[path[-1]][0])
  path.reverse()
  if len(path)<=endPoint:
    print("Path to Goal: ")
    for x in path:
      y = 0
      print("+-------+")
      for i in range(0,3):
        print("|",end=" ")
        for j in range(0,3):
          print(x[y],end=" ")
          y+=1
        print("|")
      print("+-------+")
  else: print('The length of the Path of the goal: ',len(path)-1)
  print("Cost of path: ",parent[goal][1])

def aStar(initialState,explored,frontierQ,parent):
  success=0
  frontierSQ= []  # for search only
  heapq.heappush(frontierQ ,(manhattan(changer(initialState),changer(goal)), initialState)) # insert s to heap
  heapq.heappush(frontierSQ ,initialState)
  #parent={initialstate : initialstate}  # s is parent of s
  while(frontierQ):
    cur=heapq.heappop(frontierQ)
    frontierSQ.remove(cur[1])
    cur_gn=cur[0]-manhattan(changer(cur[1]),changer(goal))+1
    if(cur[1] == goal):
      success=1
      explored.add(cur[1])
      break
    childs=create_child(cur[1],explored) # in this step make children and inset state to explored
    for ch in childs:
        if ch not in explored: 
          if ch not in frontierSQ :
            fn=cur_gn+manhattan(changer(ch),changer(goal)) # gn + hn
            heapq.heappush(frontierQ ,(fn, ch))
            heapq.heappush(frontierSQ ,ch)
            parent[ch]=[cur[1],cur_gn]
          else : # update cost and the parent if new cost < old 
              for i in range(0,len(frontierQ)):
                if ch == frontierQ[i][1]:
                    vr=frontierQ[i]
                    if(vr[0]>cur_gn+manhattan(changer(ch),changer(goal))):
                      frontierQ.remove(vr)
                      heapq.heappush(frontierQ ,(cur_gn+manhattan(changer(ch),changer(goal)), ch))
                      parent[ch]=[cur[1],cur_gn]
                    break 
  return success
def eightPuzzle_BFS(state,explored,frontierStack,parentMap,cost):
  frontierStack.append(state)
  success = 0
  while(frontierStack):
    current = frontierStack.pop(0)
    if(current == goal):
      explored.add(current)
      success = 1
      break
    childs = create_child(current,explored)
    for ch in childs:
      exit = 0
      if ch in explored:
        exit = 1
      elif ch in frontierStack:
        exit = 1
      if(exit == 1):
        continue
      frontierStack.append(ch)
      parentMap[ch]=current
      cost[ch] = cost[current] + 1
  return success

def eightPuzzle_DFS(state,explored,frontierStack,parentMap,cost):
  frontierStack.append(state)
  success = 0
  while(frontierStack):
    current = frontierStack.pop()
    if(current == goal):
      explored.add(current)
      success = 1
      break
    childs = create_child(current,explored)
    for ch in childs:
      exit = 0
      if ch in explored:
        exit = 1
      elif ch in frontierStack:
        exit = 1
      if(exit == 1):
        continue
      frontierStack.append(ch)
      parentMap[ch]=current
      cost[ch] = cost[current] + 1
  return success

def BDFSPath2Goal(initialState,parentMap,cost,endPoint):
  path = [goal]
  while path[-1] != initialState:
    path.append(parentMap[path[-1]])
  path.reverse()
  if len(path)<=endPoint:
    print("Path to Goal: ")
    for x in path:
      y = 0
      print("+-------+")
      for i in range(0,3):
        print("|",end=" ")
        for j in range(0,3):
          print(x[y],end=" ")
          y+=1
        print("|")
      print("+-------+")
  else: print('The length of the Path of the goal: ',len(path)-1)
  print("Cost of path: ",cost[goal])

def main():

  #initialState = "123046758"
  initialState = "125340678"
  #initialState = "120345678"
  #initialState = "812043765"  #el unsolvable
  #intialState = '724506831'
  #initialState='142658730'
  #initialState='125340678'
  #initialState='567408321'
  #frontierQ={initialstate : manhattan(changer(initialstate),changer(goal))}
  frontierQ= []  # list
  frontierStack = []
  frontierQQ=[]
  parent={initialState : [initialState,0]} # hash table
  parentMap = {initialState:initialState}
  parentMapD = {initialState:initialState}
  cost = {initialState:0}
  costD = {initialState:0}
  explored = set()  # set 
  exploredBFS = set() 
  exploredDFS = set() 

  start_time = time.time()
  suc = aStar(initialState,explored,frontierQ,parent)
  
  if(suc):
      print('This State is solvable! PLease Wait..... : ' + initialState)
      print('1)A * Search: ')
      print("Running Time : %f milliseconds " % ((time.time() - start_time)*1000))
      aStarPath2Goal(initialState,parent,20)
      print('Nodes Expanded: ',len(explored))
      print('Search Depth: ',getMaxDepth(parent))
      print('#######################################################')
      print('2)BFS Search: ')
      start_time = time.time()
      eightPuzzle_BFS(initialState,exploredBFS,frontierStack,parentMap,cost)
      print("Running Time : %f milliseconds " % ((time.time() - start_time)*1000))
      BDFSPath2Goal(initialState,parentMap,cost,20)
      print('Nodes Expanded: ',len(exploredBFS))
      print('Search Depth: ',max(list( cost.values() ) ))
      print('#######################################################')
      print('3)DFS Search: ')
      start_time = time.time()
      eightPuzzle_DFS(initialState,exploredDFS,frontierQQ,parentMapD,costD)
      print("Running Time : %f milliseconds " % ((time.time() - start_time)*1000))
      BDFSPath2Goal(initialState,parentMapD,costD,20)
      print('Nodes Expanded: ',len(exploredDFS))
      print('Search Depth: ',max(list( costD.values() ) ))

  else: print("This State is unsolvable: " + initialState)

main()
```
